```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
  

df = pd.read_csv('cleaned_data.csv')
```

```python
for c in ["Seychelles", "South Africa", "Brazil", "Mexico", "India", "Iran", "Russia", "United Kingdom", "Vanuatu", "Australia"]:
  df.loc[df.iloc[:,0] == c].plot(x='date', y='stringencyindex', kind="line")
  plt.title(c+" stringencyindex trend")
  plt.show()
```
