```python
import pandas as pd

df = pd.read_csv('output.csv')
df.head(n=200)
```

```python

subset_df = df[df['days_since_first'] == 375]
subset_df = subset_df.sort_values('death', ascending=False)

subset_df.head(n=5)
```

```python
subset_df = subset_df.sort_values('positive', ascending=False)
subset_df.head(n=5)
#visualize FL, PA, IL
```

```python
fl_df = df[df['state'] == 'FL']
import matplotlib.pyplot as plt
```

```python
plt.scatter(fl_df['days_since_first'], fl_df['death'])
plt.xlabel('Days since first')
plt.ylabel('Counts')
plt.title('FL Death')
plt.show()
plt.scatter(fl_df['days_since_first'], fl_df['positive'])
plt.xlabel('Days since first')
plt.ylabel('Positive')
plt.title('FL Positive')
plt.show()
```

```python
pa_df = df[df['state'] == 'PA']
plt.scatter(pa_df['days_since_first'], pa_df['death'])
plt.xlabel('Days since first')
plt.ylabel('Counts')
plt.title('PA Death')
plt.show()
plt.scatter(pa_df['days_since_first'], pa_df['positive'])
plt.xlabel('Days since first')
plt.ylabel('Positive')
plt.title('PA Positive')
plt.show()
```

```python
il_df = df[df['state'] == 'IL']
plt.scatter(il_df['days_since_first'], il_df['death'])
plt.xlabel('Days since first')
plt.ylabel('Counts')
plt.title('IL Death')
plt.show()
plt.scatter(il_df['days_since_first'], il_df['positive'])
plt.xlabel('Days since first')
plt.ylabel('Positive')
plt.title('IL Positive')
plt.show()
```
